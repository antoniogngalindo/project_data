### Importo el archivo Hechos_Homicidios.xlsx

```python
df_hechos = pd.read_excel('Hechos_Homicidios.xlsx')
```

```python
df_hechos
```

### Creo una columna Semestre

```python
df_hechos['Semestre'] = df_hechos['Mes'].apply(lambda mes: 1 if mes <= 6 else 2)
```

### Agrupo df_hechos en las columnas 'Año', 'Semestre' y 'Comuna' somando N_Victimas

```python
df_comuna = df_hechos.groupby(['Año', 'Semestre', 'Comuna'])['N_Victimas'].sum().reset_index()
```

```python
df_comuna
```

### Juntos dataframes utilizando las columnas Año y Comuna como referencia 

```python
df_comuna = pd.merge(df_comuna, df_poblacion, on= ['Año', 'Comuna'])
```

```python
df_comuna
```

### Creo una columna Tasa_Homicidios calculando la tasa de homicidio en cada linea y creo una columna Reduccion, donde uso pct_change() * 100 para calcular la reducción percentual

```python
df_comuna['Tasa_Homicidios'] = (df_comuna['N_Victimas'] / df_comuna['Población']) * 100000

df_comuna['Reduccion'] = df_comuna.groupby(['Comuna'])['Tasa_Homicidios'].pct_change() * 100
```

```python
df_comuna
```

```python
df_comuna.isna().sum()
```

### Cambio los valores de NA en la columna Reducción porque son los primeros semestres de 2016, luego como no hay datos de 2015, no hubo calculo

```python
df_comuna.fillna(0, inplace= True)
```
